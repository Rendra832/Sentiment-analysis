```python
import pandas as pd
from youtube_comment_downloader import YoutubeCommentDownloader
```

```python
def scrape_comments_from_urls(url_list, max_comments_total=5000):
    downloader = YoutubeCommentDownloader()
    comments = []
    for url in url_list:
        count = 0
        print(f"Scraping: {url}")
        for c in downloader.get_comments_from_url(url):
            text = c.get('text')
            if text:
                comments.append({'comment': text, 'video': url})
                count += 1
            if len(comments) >= max_comments_total:
                break
        print(f" -> Collected {count} from this video, total so far {len(comments)}")
        if len(comments) >= max_comments_total:
            break
    return pd.DataFrame(comments)
```

```python
urls = [
    "https://youtu.be/NIr72bFGFts?si=YHGohGGVPQ91ZM40",
]

df = scrape_comments_from_urls(urls, max_comments_total=5000)

print("Total comments scraped:", len(df))
```

```python
df.to_csv("youtube_comments.csv", index=False)
print("File tersimpan sebagai youtube_comments.csv")
```
